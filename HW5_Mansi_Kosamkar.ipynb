#### Step 1.1 - Read in the data. Create a table of the unique values of Category. Which values do you believe represent a restaurant? Explain which categories you chose in a markdown cell. Create a single dummy variable for restaurant that combines multiple values from Category. 

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df1 = pd.read_csv("Food_Inspection.csv")
df1
```

```python
#finding unique types
unique_types = df1.Category.unique()
```

```python

df_unique = pd.DataFrame(unique_types)
df_unique
```

```python
#getting dummy values
dummies = pd.get_dummies(df1['Category'])
df_all = pd.concat([df1,dummies],axis=1)
```

```python
df_all['Combined'] = df_all['Restaurant'] + df_all['Snack Bar'] 
df_all.head(15)
```

I used the categories of restaurant and snackbar as I beleive they belong to the category of restaurants

#### Step 1.2 - Convert the Inspection_date column into a datetime column. Create a new column for the year of the inspection. Create a new column for the month of the inspection. Create a column for the year and month.

```python

df_all['Inspection Date']= pd.to_datetime(df1['Inspection Date']) 
df_all['Inspection Year']= df_all['Inspection Date'].dt.year
df_all['Inspection Month']= df_all['Inspection Date'].dt.month
df_all['Inspection Year and Month']= df_all['Inspection Date'].dt.to_period('M')
df_all.head(15)
```

#### Step 1.3 - For each column with the type of compliance, e.g. "Rodent and Insects," create a dummy variable that is 1 if the establishment is out of compliance and 0 otherwise. Use np.nan for not applicable. Drop the string columns (retain only the dummy variables).

```python
sub = df1[['Food from Approved Source (C)','Food Protected from Contamination (C)', 'Ill Workers Restricted (C)','Proper Hand Washing (C)', 'Cooling Time and Temperature (C)','Cold Holding Temperature (C)','Hot Holding Temperature (C)','Cooking Time and Temperature (C)','Reheating Time and Temperature (C)','Hot and Cold Running Water Provided (C)','Proper Sewage Disposal (C)','Toxic Substances & Pesticides','Rodent and Insects', 'Nutritional Labeling','Trans Fat Ban', 'No-Smoking Sign Posted']]
d_col = sub.copy()
l = LabelEncoder()


columns = ['Food from Approved Source (C)', 'Food Protected from Contamination (C)', 'Ill Workers Restricted (C)', 'Proper Hand Washing (C)', 'Cooling Time and Temperature (C)', 'Cold Holding Temperature (C)', 'Hot Holding Temperature (C)', 'Cooking Time and Temperature (C)', 'Reheating Time and Temperature (C)', 'Hot and Cold Running Water Provided (C)', 'Proper Sewage Disposal (C)', 'Toxic Substances & Pesticides', 'Rodent and Insects', 'Nutritional Labeling', 'Trans Fat Ban', 'No-Smoking Sign Posted']

for column in columns :
    d_col[column] = l.fit_transform(d_col[column])

d_col['Nutritional Labeling'] = np.where(np.array(d_col['Nutritional Labeling']) == 1, np.nan, 0)
df2 = d_col
d_col.head(15)
```

#### Step 1.4 - Create a new column that contains the number of violations for that inspection (the number of categories where the establishment was not in compliance). Create a dummy variable that is 1 if the establishment is out of compliance in any category. 

```python
df2['Violations per Inspection'] = d_col.sum(axis = 1).values
df2.head(15)
```

```python
df2['Compliancy'] = df2.sum(axis = 1, skipna = True)
df2['Compliancy'] = np.where(df2['Compliancy'] > 0, 1, 0)
df2.head(15)
```

```python
df_all = pd.concat([df_all,df2[['Violations per Inspection','Compliancy']]], axis=1)
```

Step 1.5 - For establishments with multiple inspections, create a new DataFrame in wide format. Keep only the establishment ID, Category, Inspection_date, and number of violations. Make sure category is consistent within ID and resolve any discrepancies if necessary (i.e. each establishment has only one category). Reshape from long to wide (pivot) such that each establishment is a row and you have a column for the date and number of violations for inspection 1, inspection 2, inspection 3, etc. 

```python
new_cols=['Establishment ID','Category','Inspection Date','Violations per Inspection']
df_reshape = df_all[new_cols]
df_reshape  = df_reshape [df_reshape['Establishment ID'].isin(df_reshape['Establishment ID'].value_counts()[df_reshape['Establishment ID'].value_counts()>1].index)]
df_reshape=df_reshape.sort_values(by=['Establishment ID','Inspection Date'])
df_reshape .head()
```

```python
df_reshape_new = df_reshape.pivot_table(index='Establishment ID', values=['Inspection Date','Violations per Inspection'], aggfunc=lambda x:' '.join(str(v) for v in x), columns=['Inspection Date'])
df_reshape_new = df_reshape_new.swaplevel(0,1,axis=1).sort_index(axis=1).fillna('-')
df_reshape_new
```

#### Step 3.2 - Create a line graph that shows the percent of restaurant inspections that have at least one violation by month and year. Are inspections getting harder or easier over time? Is there a particular month where more restaurants pass? Write 2-4 sentences with your observations.

```python
final = df_all[df_all['Restaurant']==1]
m_y =final.groupby('Inspection Year and Month')

date_of_inspection = {'Inspection Year and Month':[], '%':[]}

for date, patch in m_y:
  date_of_inspection['Inspection Year and Month'].append(str(date.month) + '-' + str(date.year))
  temp = patch[patch['Compliancy']==1]
  t = temp.shape[0]
  tot = patch.shape[0]
  date_of_inspection['%'].append(t/tot * 100)

date_of_inspection = pd.DataFrame(date_of_inspection)
date_of_inspection.set_index('Inspection Year and Month', inplace=True)

fig, ax = plt.subplots(figsize=(10,7))
plt.ylabel('Non-Compliant Restaurants')
plt.xlabel('Month - Year')
plt.title('Non-Compliant Inspections(Month and Year)')
date_of_inspection.plot(ax=ax)
```

The number of violations in non - compliant restaurants took a rise in 2017, 2018 and towards the end of 2019

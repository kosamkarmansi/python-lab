```python
poke_url = 'https://pokemondb.net/pokedex/all'
# Issue a simple HTTP request to get the webpage text
poke_page = requests.get(poke_url)
# Response code is returned
poke_page
```

```python
soup = BeautifulSoup(poke_page.text,'html.parser') 
```

```python
table = soup.find(name ="table", class_="data-table")

table_full = table.find("tbody")
table_rows = table_full.find_all("tr")

table_full = table.find("thead")
table_rows_head = table_full.find_all("tr")

```

Step 1.3: Generalize step 1.2. Define a function that takes in a row of the pokedex table and returns it as a DataFrame with a single row. Create a single DataFrame by appending these rows. (Appending the data to a list or dictionary and then creating a DataFrame is also acceptable.) Make sure you skip the header row.

```python
def poke_mon(inp_row):
    poke_list = []
    poke_list.append(table_rows[inp_row].find_all(name = "td", class_ = "cell-num")[0].text)
    poke_list.append(table_rows[inp_row].find(name = "a", class_= "ent-name").text)
    poke_list.append(table_rows[inp_row].find(name = "a", class_= "ent-name")['href'])
    poke_list.append(table_rows[inp_row].find(name = "td", class_= "cell-icon").text)
    poke_list.append(table_rows[inp_row].find(name = "td", class_= "cell-total").text)
    
    n = table_rows[inp_row].find_all(name = 'td', class_= "cell-num")
    
    for i in range(1,7):
        poke_list.append(n[i].text)
        
    poke_dex = pd.DataFrame(poke_list,index=["ID Number", "Name", "URL", "Type", "Total Points", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"])
    
    poke_dex = poke_dex.T
    return poke_dex
    
```

```python
poke_dataframe = []
for j in range(rows_count-1):
    result = poke_mon(j)
    poke_dataframe.append(result)
poke_df = pd.concat(poke_dataframe)
poke_df

```
